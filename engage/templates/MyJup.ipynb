```python
carsale.replace({'engType': 'Other', 'price': 0, 'mileage': 0}, np.nan, inplace=True)
limitPer = len(carsale) * .80
carsale = carsale.dropna(thresh=limitPer, axis=1)
```

```python
profile = pandas_profiling.ProfileReport(carsale)
profile.to_file("carsale_before_preprocessing_2.html")
```

```python
carsale.drop_duplicates(inplace=True) #inplace used to modify the dataset with applied command
carsale.shape
```

```python
def get_median_price(x):
    brand = x.name[0]
    if x.count() > 0:
        return x.median() # Return median for a brand/model if the median exists.
    elif carsale.groupby(['Make'])['Price'].count()[brand] > 0:
        brand_median = carsale.groupby(['Make'])['Price'].apply(lambda x: x.median())[brand]
        return brand_median # Return median of brand if particular brand/model combo has no median,
    else:                 # but brand itself has a median for the 'price' feature. 
        return carsale['Price'].median() # Otherwise return dataset's median for the 'price' feature.

price_median = carsale.groupby(['Make','Model'])["Price"].apply(get_median_price).reset_index()
price_median.rename(columns={'Price': 'price_med'}, inplace=True)
price_median.head()
```

```python
carsale.columns
```

```python
carsale['Price'] = carsale['Price'].str.replace(',', '').astype(float)
carsale['Price'] = carsale['Price'].astype(float)
```

```python
carsale['Price'] = carsale['Price'].str.replace('Rs. ', '')
```

```python
type(carsale['Price'][0] )
```

```python
def fill_with_median(x):
    if pd.isnull(x['Price']):
        return price_median[(price_median['Make'] == x['Make']) & (price_median['Model'] == x['Model'])]['price_med'].values[0]
    else:
        return x['Price']
    
carsale['Price'] = carsale.apply(fill_with_median, axis=1)
carsale.head()
```

```python
def get_median_engV(x):
    brand = x.name[0]
    if x.count() > 0:
        return x.median() # Return median for a brand/model if the median exists.
    elif carsale.groupby(['Make'])['Cylinders'].count()[brand] > 0:
        brand_median = carsale.groupby(['Make'])['Cylinders'].apply(lambda x: x.median())[brand]
        return brand_median # Return median of brand if particular brand/model combo has no median,
    else:                 # but brand itself has a median for the 'engV' feature. 
        return carsale['Cylinders'].median() # Otherwise return dataset's median for the 'engV' feature.
    
engV_median = carsale.groupby(['Make','Model'])['Cylinders'].apply(get_median_engV).reset_index()
engV_median.rename(columns={'Cylinders': 'engV_med'}, inplace=True)
engV_median.head()
```

```python
def fill_with_median(x):
    if pd.isnull(x['Cylinders']):
        return engV_median[(engV_median['Make'] == x['Make']) & (engV_median['Model'] == x['Model'])]['engV_med'].values[0]
    else:
        return x['Cylinders']
    
carsale['Cylinders'] = carsale.apply(fill_with_median, axis=1)
carsale.head()
```

```python
def get_drive_mode(x):
    brand = x.name[0]
    if x.count() > 0:
        return x.mode() # Return mode for a brand/model if the mode exists.
    elif carsale.groupby(['Make'])['Drivetrain'].count()[brand] > 0:
        brand_mode = carsale.groupby(['Make'])['Drivetrain'].apply(lambda x: x.mode())[brand]
        return brand_mode # Return mode of brand if particular brand/model combo has no mode,
    else:                 # but brand itself has a mode for the 'drive' feature. 
        return carsale['drive'].mode() # Otherwise return dataset's mode for the 'drive' feature.
    
drive_modes = carsale.groupby(['Make','Model'])['Drivetrain'].apply(get_drive_mode).reset_index().drop('level_2', axis=1)
drive_modes.rename(columns={'Drivetrain': 'drive_mode'}, inplace=True)
drive_modes.head()
```
